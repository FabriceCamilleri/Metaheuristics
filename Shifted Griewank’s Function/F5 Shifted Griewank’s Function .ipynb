```python
class F5_Shifted_Griewank:    

    nbFitness = 0
    
    def __init__(self, dim, bounds):
        self.dim = dim
        self.bounds=bounds
        F5_Shifted_Griewank.nbFitness = 0

    def fitness(self, x):
        F1 = 0
        F2 = 1
        for i in range(0,x.shape[0]):    
            z = x[i] - griewank[i]
            F1 = F1 + ( z**2 / 4000 )
            F2 = F2 * ( np.cos(z/np.sqrt(i+1)))
            F5_Shifted_Griewank.nbFitness += 1
        return F1 - F2 + 1 + f_bias[4]

```
