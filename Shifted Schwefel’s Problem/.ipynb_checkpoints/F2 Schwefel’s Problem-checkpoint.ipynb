```python
class F2_Schwefel_Problem:
    
    nbFitness = 0
    
    def __init__(self, dim, bounds):
        self.dim = dim
        self.bounds=bounds

    def fitness(self, x):
        F = abs(x[0])
        for i in range(0,x.shape[0]):
            z = x[i] - schwefel[i]
            F = np.maximum(F , abs(z))
        return [F + f_bias[1]]

    def get_bounds(self):
        return ([-100] * self.dim, [100] * self.dim)
    
    def get_name(self):
        return "F2: Schwefel Problem "

    def get_extra_info(self):
        return "\tDimensions: " + str(self.dim)
```

```python
prob = pg.problem(F2_Schwefel_Problem(50))
```
