```python
class F6_Shifted_Ackley:    

    nbFitness = 0
    
    def __init__(self, dim, bounds):
        self.dim = dim
        self.bounds=bounds
        F6_Shifted_Ackley.nbFitness = 0

    def fitness(self, x):
        Sum1 = 0
        Sum2 = 0
        F = 0
        for i in range(0,x.shape[0]):  
            z = x[i] - ackley[i]
            Sum1 = Sum1 + z**2
            Sum2 = Sum2 + np.cos(2*np.pi*z)
        F = -20*np.exp(-0.2*np.sqrt(Sum1/x.shape[0])) -np.exp(Sum2/x.shape[0]) + 20 + np.e + f_bias[5]
        F6_Shifted_Ackley.nbFitness += 1
        return F 
```
